```python
import random

def populate_list(length):
    """
    Populates a list of the given length with random integers between 10 and 50.
    """
    lst = []
    for i in range(length):
        lst.append(random.randint(10, 50))
    return lst

def sum_list(lst):
    """
    Sums all the elements of the given list.
    """
    return sum(lst)

# Ask the user to enter a number between 5 and 15
user_input = int(input("Enter an integer number between 5 and 15: "))

# Check if the user input is between 5 and 15
if user_input >= 5 and user_input <= 15:
    # Populate a list with random integers
    my_list = populate_list(user_input)
    # Display the list in the console
    print("List: ", my_list)
    # Sum all the elements of the list
    total = sum_list(my_list)
    # Display the sum in the console
    print("Sum: ", total)
else:
    print("Invalid input. Please enter a number between 5 and 15.")
```
